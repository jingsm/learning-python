## Visualizing with Seaborn

```python
import seaborn as sbn
%matplotlib inline
```

```python
tips = sbn.load_dataset('tips')
```

```python
tips.head()
```

```python
sbn.distplot(tips['total_bill'])
```

```python
sbn.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sbn.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sbn.pairplot(tips)
```

```python
sbn.pairplot(tips,hue='sex',palette='husl')
```

```python
sbn.barplot(x='sex',y='total_bill',data=tips)
```

```python
sbn.countplot(x='sex',data=tips)
```

```python
sbn.boxplot(x='day',y='total_bill',data=tips)
```

```python
sbn.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='rainbow')
```

```python
sbn.stripplot(x='day',y='total_bill',data=tips)
```

```python
sbn.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sbn.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1')
```
